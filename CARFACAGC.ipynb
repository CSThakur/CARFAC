CAR FAC - Cascade of Asymmetric Resonators with Fast Acting Compression
===

The code below implements a model of the cochlea developed by Dick Lyon.

It will be described in an upcoming book by him. Please contact him directly at <dicklyon@acm.org> for more information.


Following on from the previous notebook on [the CARFAC model](http://nbviewer.ipython.org/github/vschaik/CARFAC/blob/master/CARFAC.ipynb) which added the Outer Hair Cell to the Basilar Membrane model, we now incorporate the feedback loop that controls the gain of the OHCs. To do so, we add 4 temporal low pass filters to the output of the Inner Hair Cells, which we then couple with their neighbours to also implement a spatial low pass filter. The new structure looks like this:

<img src="images/AGC.png " width=600>

To measure the linearised transfer functions of the adapted cochlea, we first play a stimulus through the cochlea, after which we fix the value of b for each section, and then run the log sine sweep signal through the cochlea as before to measure the transfer function.


```python
from pylab import *
from scipy import signal
import colorsys
from mpl_toolkits.mplot3d import Axes3D

fs = 32000.0                  # sample frequency
dur = 1.0                     # stimulus length
npoints = int(fs*dur)

nsec = 100                    # number of sections in the cochlea between
xlow = 0.1                    # lowest frequency position along the cochlea and
xhigh = 0.90                  # highest frequency position along the cochlea

# create input tone
f0 = 1e3                      # tone frequency
t1 = arange(npoints)/fs       # sample times
gain = 0.1                    # input gain
stimulus = gain*sin(2*pi*f0*t1)

# BM parameters
x = linspace(xhigh,xlow,nsec) # position along the cochlea 1 = base, 0 = apex
f = 165.4*(10**(2.1*x)-1)     # Greenwood for humans
a0 = cos(2*pi*f/fs)           # a0 and c0 control the poles and zeros
c0 = sin(2*pi*f/fs)

damping = 0.4                 # damping factor
r = 1 - damping*2*pi*f/fs     # pole & zero radius actual
r1 = 1 - damping*2*pi*f/fs    # pole & zero radius minimum (set point)
h = c0                        # p273 h=c0 puts the zeros 1/2 octave above poles
#h=0.5*(2+2*a0)/c0             # see Fig 15.3 and p273
g = (1-2*a0*r+r*r)/(1-(2*a0-h*c0)*r+r*r)  # p273 this gives 0dB DC gain for BM

f_hpf = 20                    # p294 20Hz corner for the BM HPF
q = 1/(1+(2*pi*f_hpf/fs))     # corresponding IIR coefficient 

# IHC parameters
z = 10                        # p295
v_max = z**3/(z**3+z**2+0.1)  # p295 maximum v_mem signal
tau_in = 10e-3                # p295 transmitter creation time constant
tau_out = 0.5e-3              # p295 transmitter depletion time constant
c_in = 1/(fs*tau_in*v_max)    # p295 corresponding IIR coefficient    
c_out = 1/(fs*tau_out)        # p295 corresponding IIR coefficient 
tau_IHC = 80e-6               # p295 ~2kHz LPF for IHC output
c_IHC = 1/(fs*tau_IHC)        # corresponding IIR coefficient 

# OHC parameters
scale = 0.1                   # p281
offset = 0.04                 # p281 
b = 0.5*ones(nsec)            # automatic gain loop feedback (1=no undamping).
d_rz = 0.7*(damping*2*pi*f/fs)# OHC gain

# AGC loop parameters
# temporal
tau_AGC1 = .128
tau_AGC2 = .032
tau_AGC3 = .008
tau_AGC4 = .002

c_AGC1 = 1/(fs*tau_AGC1)
c_AGC2 = 1/(fs*tau_AGC2)
c_AGC3 = 1/(fs*tau_AGC3)
c_AGC4 = 1/(fs*tau_AGC4)

# spatial
sa = 0.286
sb = 0.404
sc = 1 - sa -sb


W0 = zeros(nsec)              # BM filter internal state
W1 = zeros(nsec)              # BM filter internal state
W1old = zeros(nsec)           # BM filter internal state at t-1
BM = zeros((nsec,npoints))    # BM displacement
BM_hpf = zeros((nsec,npoints))# BM displacement high-pass filtered at 20Hz
trans = ones(nsec)            # transmitter available
IHC = zeros((nsec,npoints))   # IHC output
IHCa = zeros((nsec,npoints))  # IHC filter internal state
AGC = zeros((nsec,npoints))   # AGC filter internal state
AGC1 = zeros(nsec)            # AGC filter internal state
AGC2 = zeros(nsec)            # AGC filter internal state
AGC3 = zeros(nsec)            # AGC filter internal state
AGC4 = zeros(nsec)            # AGC filter internal state


#%% play through cochlea
BM[-1] = stimulus             # put stimulus at BM[-1] to provide input to BM[0]
BM[-1,-1] = 0                 # hack to make BM_hpf[nsec-1,0] work
for t in range(npoints):
    for s in range(nsec):
        W0new = BM[s-1,t] + r[s]*(a0[s]*W0[s] - c0[s]*W1[s])
        W1[s] = r[s]*(a0[s]*W1[s] + c0[s]*W0[s])
        W0[s] = W0new
        BM[s,t] = g[s]*(BM[s-1,t] + h[s]*W1[s])
    BM_hpf[:,t] = q*(BM_hpf[:,t-1] + BM[:,t] - BM[:,t-1])
    z_int = (1 - (BM_hpf[:,t]+0.13)/4).clip(0)
    z = ((z_int)**8).clip(0,1)
    v_mem = 0.75*((1-z)**2)
    IHC_new = v_mem*trans    
    trans += c_in*(1-trans) - c_out*IHC_new   
    IHCa[:,t] = (1-c_IHC)*IHCa[:,t-1] + c_IHC*IHC_new
    IHC[:,t] = (1-c_IHC)*IHC[:,t-1] + c_IHC*IHCa[:,t]
    v_OHC = W1 - W1old
    sqr=(v_OHC*scale+offset)**2
    NLF= 1/(1 + (scale*v_OHC + offset)**2)
    r = r1 + d_rz*(1-b)*NLF
    g = (1-2*a0*r+r*r)/(1-(2*a0-h*c0)*r+r*r)
    AGC1 = (1-c_AGC1)*AGC1 + c_AGC1*IHC[:,t]
    AGC2 = (1-c_AGC1)*AGC2 + c_AGC2*IHC[:,t]
    AGC3 = (1-c_AGC1)*AGC3 + c_AGC3*IHC[:,t]
    AGC4 = (1-c_AGC1)*AGC4 + c_AGC4*IHC[:,t]
    AGC[:,t] = AGC1 + AGC2 + AGC3 + AGC4
    AGC[:,t] = sa*(append(AGC[1:,t],AGC[-1,t])) + sc*AGC[:,t] + sb*(append(AGC[0,t],AGC[0:-1,t]))
    b = AGC[:,t]/4.0
    b = b.clip(0,1)

#%%
# create a log-sine-sweep
f0 = 10                       # sweep start frequency
f1 = fs/2                     # sweep end frequency
stimulus = gain*signal.chirp(t1, f0, t1[-1], f1, method='logarithmic', phi=-90)

#%% now play sweep through cochlea while keeping b fixed to measure the frequency response
BM[-1] = stimulus             # put stimulus at BM[-1] to provide input to BM[0]
BM[-1,-1] = 0                 # hack to make BM_hpf[nsec-1,0] work
for t in range(npoints):
    for s in range(nsec):
        W0new = BM[s-1,t] + r[s]*(a0[s]*W0[s] - c0[s]*W1[s])
        W1[s] = r[s]*(a0[s]*W1[s] + c0[s]*W0[s])
        W0[s] = W0new
        BM[s,t] = g[s]*(BM[s-1,t] + h[s]*W1[s])
    BM_hpf[:,t] = q*(BM_hpf[:,t-1] + BM[:,t] - BM[:,t-1])
    z_int = (1 - (BM_hpf[:,t]+0.13)/4).clip(0)
    z = ((z_int)**8).clip(0,1)
    v_mem = 0.75*((1-z)**2)
    IHC_new = v_mem*trans    
    trans += c_in*(1-trans) - c_out*IHC_new   
    IHCa[:,t] = (1-c_IHC)*IHCa[:,t-1] + c_IHC*IHC_new
    IHC[:,t] = (1-c_IHC)*IHC[:,t-1] + c_IHC*IHCa[:,t]
    v_OHC = W1 - W1old
    sqr=(v_OHC*scale+offset)**2
    NLF= 1/(1 + (scale*v_OHC + offset)**2)
    r = r1 + d_rz*(1-b)*NLF
    g = (1-2*a0*r+r*r)/(1-(2*a0-h*c0)*r+r*r)


#%%
# play the stimulus through the system and measure the output
output = append(zeros((nsec,1)),BM_hpf,1)
stimulus = append(0,stimulus)

# use the FFT of the stimulus and output directly to calculate the transfer function
myFFT = fft(zeros((nsec,(npoints+1))))
for s in range(nsec):
    myFFT[s] = (fft(output[s])/fft(stimulus))

# calculate the BM impulse response
IR = real(ifft(myFFT))
IR[:,0] = zeros(nsec)   #remove artefact

```

First we plot the same data as before: BM gain and phase response, impulse response and IHC output in response to the sweep:

```python
# plot the data

figure(1, figsize=(16,4), dpi=300) # Bode plot of BM displacement 
ax1 = subplot(1,2,1)
freq = linspace(0,fs/2,(npoints+1)/2)
semilogx(freq,20*log10(abs(myFFT.T[0:(npoints+1)/2,:])))
title('BM gain (in dB)')
ax2 = subplot(1,2,2,sharex=ax1)
semilogx(freq,unwrap(angle(myFFT.T[0:(npoints+1)/2,:]),discont=pi,axis=0))
ylim([-20,2])
xlabel('f (in Hz)')
title('BM phase (in rad)')

figure(2, figsize=(16,3), dpi=300) # BM impulse response
L=2000
plot(arange(L)*1000/fs,IR[:,0:L].T)
xlabel('t (in ms)')
title('BM impulse response')

figure(3, figsize=(16,3), dpi=300) # IHC output
plot(t1*1000,stimulus[:-1],'r')
plot(t1*1000,IHC.T)
xlabel('t (in ms)')
title('IHC response')

```

Next we plot the output of the AGC loop versus section number and time. The top figure shows the 3d version of this, and the bottom figure the cross sections. You can see that the output becomes stable after 0.4s of the presentation of the pure tone.

```python
fig = figure(4, figsize=(8,6), dpi=300)
ax = fig.add_subplot(111, projection='3d')
t = linspace(0,1,10,False)*dur
X,Y = meshgrid(linspace(0,nsec,nsec,False),t)
ax.plot_surface(X,Y,AGC[:,(t*fs).astype(int)].T, rstride=10, cstride=10, cmap=cm.jet, linewidth=0)
ax.set_xlabel("section number")
ax.set_ylabel("Time (s)")
ax.set_zlabel("AGC output")

figure(5, figsize=(8,4), dpi=300)
for t in range(10):
    plot(AGC[:,t*npoints/10], label="%0.2f" % (t*npoints/10/fs), color=colorsys.hls_to_rgb(t/10.0, .5, .5))
legend()
title ('AGC output')
xlabel('section number')

```
